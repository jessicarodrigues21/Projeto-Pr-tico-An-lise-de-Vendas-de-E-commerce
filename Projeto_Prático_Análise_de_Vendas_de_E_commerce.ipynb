```python
#Coluna Idade
#Media, Mediana e Desvio Padrão
media_idade = df['Idade'].mean()
print(f'Média das idade {media_idade:.1f}')

mediana_idade = df['Idade'].median()
print(f'Mediana das idades {mediana_idade}')

desvio_padrao_idade = df['Idade'].std()
print(f'Desvio padrão das idades {desvio_padrao_idade:.1f}')
```

```python
#Coluna Compras
#Media, Mediana e Desvio Padrão
media_compras = df['Valor Compra'].mean()
print(f'Média das compras {media_compras:.1f}')

mediana_compras = df['Valor Compra'].median()
print(f'Mediana das compras {mediana_compras:.1f}')

desvio_padrao_compras = df['Valor Compra'].std()
print(f'Desvio padrão das idades {desvio_padrao_compras:.1f}')
```

**3. Manipulação e Limpeza dos Dados:**

**Crie novas colunas que possam ser úteis para a análise, como faixa etária ou categorias simplificadas.**

**Limpe e transforme os dados conforme necessário para facilitar a análise.**

```python
#Faixa Etária
faixas_etarias = [0, 18, 25, 35, 50, 100]
rotulo_faixa_etaria = ['0 - 18', '19 - 25', '26 - 35', '36 - 50', '51 +']
df['Faixa Etária'] = pd.cut(df['Idade'], bins=faixas_etarias, labels=rotulo_faixa_etaria, right=False)

df
```

**4. Análise Detalhada:**

**Identifique os itens mais vendidos e as categorias mais populares.**

**Analise o valor médio das compras por idade e cidade.**

**Investigue as preferências de método de pagamento entre diferentes faixas etárias.**

```python
itens = df.groupby('Item Comprado').size()
categorias = df.groupby('Categoria').size()

print(itens,'\n')
print(categorias)
```

```python
media_compras_idade = df.groupby('Idade')['Valor Compra'].mean()
print(media_compras_idade)
```

```python
media_compras_cidade = df.groupby('Cidade')['Valor Compra'].mean()
print(media_compras_cidade)
```

```python
df['Faixa Etária'] = df['Faixa Etária'].astype('category')
df['Método Pagamento'] = df['Método Pagamento'].astype('category')

# Contando as preferências de pagamento por faixa etária
preferencias_pagamento = df.groupby('Faixa Etária')['Método Pagamento'].value_counts().unstack(fill_value=0)
print(preferencias_pagamento)
```
